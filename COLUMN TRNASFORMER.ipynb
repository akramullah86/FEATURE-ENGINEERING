```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder,OrdinalEncoder

from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('E:\Datasets\covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=['has_covid'])

y = df.has_covid
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=2)
```

```python
col_transformed = ColumnTransformer(transformers = [
    
    ('imputer',SimpleImputer(),['fever']),
    
    ('oe',OrdinalEncoder(categories=[['Strong','Mild']]),['cough']),
    
    ('ohe',OneHotEncoder(drop='first',sparse=False),['gender','city']),
    
    
    
],remainder = 'passthrough')
```

```python
x_train.head()
```

```python
col_transformed.fit(x_train,y_train)
x_train_trf = col_transformed.transform(x_train)

x_train_trf.shape

pd.DataFrame(x_train_trf)
```

```python
x_test_trf = col_transformed.transform(x_test)

pd.DataFrame(x_test_trf)
```
