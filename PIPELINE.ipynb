```python
import pandas as pd

def create_category_columns(df, column_name):
    """
    Create new columns for each category in the specified column using pandas assign method.

    Args:
    df (DataFrame): The DataFrame containing the column.
    column_name (str): The name of the column containing the categories.

    Returns:
    DataFrame: The DataFrame with new columns for each category.
    """
    # Get the unique categories in the column
    categories = df[column_name].unique()

    # Create new columns for each category
    for category in categories:
        df = df.assign(**{f"{column_name}_{category}": (df[column_name] == category).astype(int)})

    return df

# Example usage
data = {'Sex': ['male', 'female', 'female', 'female', 'male'],
        'Embarked': ['S', 'C', 'S', 'S', 'S']}
df = pd.DataFrame(data)

# Create new columns for each category in the 'Sex' column
result_df = create_category_columns(df, 'Sex')

# Create new columns for each category in the 'Embarked' column
result_df = create_category_columns(result_df, 'Embarked')

print(result_df)
```
