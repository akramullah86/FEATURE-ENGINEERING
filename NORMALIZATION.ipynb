```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('E:\\Datasets\\wine_data.csv',header=None,usecols = [0,1,2])

df.columns=['class label','alchohal','malic acid']

df.head()
```

```python
x = df.iloc[:,-2:]

y = df.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=2)


x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
sc = MinMaxScaler()
```

```python
sc.fit(x_train)

x_train_scaled = sc.transform(x_train)

x_test_scaled = sc.transform(x_test)

print(x_train_scaled)

```

```python
print(x_test_scaled)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = x_train.columns)

x_test_scaled = pd.DataFrame(x_test_scaled,columns = x_test.columns)
```

```python
x_train_scaled.head()
```

```python
x_test_scaled.head()
```

## Scatter Plot befor and after Scaling

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2,figsize=(14,6))

ax1.set_title('Before Scaling')
ax1.scatter(x_train['alchohal'],x_train['malic acid'],c=y_train)


ax2.set_title('After Scaling')

ax2.scatter(x_train_scaled['alchohal'],x_train_scaled['malic acid'],c=y_train)
```

## Distribuitions Before and After Scaling

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(16,6))

sns.kdeplot(x_train['alchohal'],ax=ax1,color='black')

sns.kdeplot(x_train_scaled['alchohal'],ax=ax2,color='yellow')

ax1.set_title('Before Scaling')

ax2.set_title('After Scaling')
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(16,6))

sns.kdeplot(x_train['malic acid'],ax=ax1,color='black')

sns.kdeplot(x_train_scaled['malic acid'],ax=ax2,color='yellow')

ax1.set_title('Before Scaling')

ax2.set_title('After Scaling')
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(16,6))

# Before Scaling

sns.kdeplot(x_train['alchohal'],ax=ax1)

sns.kdeplot(x_train['malic acid'],ax=ax1)

# After scaling

sns.kdeplot(x_train_scaled['alchohal'],ax=ax2)

sns.kdeplot(x_train_scaled['malic acid'],ax=ax2)

ax1.set_title('Before Scaling')

ax2.set_title(('After Scaling'))
```
